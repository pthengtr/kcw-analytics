```python
from datetime import datetime
import zipfile
import os

suffix = datetime.now().strftime("%Y-%m-%d-%H%M")

folder_to_zip = f"{BASE_FOLDER}/KCW-Data/kcw_analytics/04_outputs"
zip_path = f"{BASE_FOLDER}/KCW-Data/kcw_analytics/06_archive/export_{suffix}.zip"

exclude_files = {"desktop.ini", "Thumbs.db", ".DS_Store"}

with zipfile.ZipFile(zip_path, "w", zipfile.ZIP_DEFLATED) as z:
    for root, dirs, files in os.walk(folder_to_zip):
        for file in files:
            if file in exclude_files:
                continue

            full_path = os.path.join(root, file)
            arcname = os.path.relpath(full_path, folder_to_zip)
            z.write(full_path, arcname)

print(f"Created: {zip_path}")
```
