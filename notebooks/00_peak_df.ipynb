```python
import pandas as pd

def make_peak_df(n=10):
    rows = []
    for i in range(1, n + 1):
        code = f"{10000000 + i:08d}"
        rows.append({
            "ลำดับที่*": i,
            "รหัสสินค้า": code,
            "ชื่อสินค้า*": f"ชื่อสินค้าทดลอง_{i:05d}",
            "คำอธิบาย": f"สินค้าทดลอง {i}",
            "ชื่อหน่วย": "ชิ้น",
            "ราคาซื้อ/หน่วย": 100 + i,
            "อัตราภาษีซื้อ": "7%",
            "ราคาขาย/หน่วย": 150 + i,
            "อัตราภาษีขาย": "7%",
            "ผังบัญชีซื้อสินค้า": "",
            "ผังบัญชีรายได้ขาย": "",
            "คำนวณต้นทุนขายหรือไม่": "คำนวณ",
            "วิธีการคำนวณ": "FIFO",
            "ผังบัญชีต้นทุนขาย": "",
            "วันที่ซื้อ (yyyymmdd)": "",
            "จำนวนหน่วย": "",
            "ราคาต่อหน่วย": "",
            "Shopee": "",
            "Lazada": "",
            "Tiktok": "",
            "บาร์โค้ด": code
        })
    return pd.DataFrame(rows)

df = make_peak_df(10)
print(df.head())
```

```python
df.to_csv(
    "peak_test_10.csv",
    index=False,
    encoding="utf-8-sig"  # VERY IMPORTANT for Thai
)
```
