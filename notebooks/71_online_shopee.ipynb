```python
import os
import pandas as pd

data = {}

for file in os.listdir(folder):
    if file.endswith(".csv"):
        path = os.path.join(folder, file)
        data[file] = pd.read_csv(
            path,
            dtype={
              "BCODE": "string",
              "MOBILE": "string",
              "BILLNO": "string",
            },
            encoding="utf-8-sig",
            low_memory=False   # stops chunk guessing
        )
        print(f"Loaded: {file} -> {data[file].shape}")
```

```python
folder = f"{BASE_FOLDER}/KCW-Data/kcw_analytics/01_raw/online/shopee"
```

```python
import pandas as pd
import glob
import os

files = glob.glob(os.path.join(folder, "*.xlsx"))

df_shopee_raw = pd.concat(
    (
        pd.read_excel(
            f,
            sheet_name="Income",
            skiprows=5,
            dtype={
              "รหัสสินค้า": "string",     # or the order id column
              "หมายเลขคำสั่งซื้อ": "string"
    }
        )
        for f in files
    ),
    ignore_index=True
)
```

```python
df_shopee_raw.columns
```

```python
df_shopee_raw[['วันที่ทำการสั่งซื้อ','หมายเลขคำสั่งซื้อ', 'รหัสคืนสินค้า', 'สินค้าราคาปกติ','จำนวนเงินทั้งหมดที่โอนแล้ว (฿)']]
```

```python
df = df_shopee_raw.copy()

df = df.rename(columns={
    'วันที่ทำการสั่งซื้อ': 'date',
    'หมายเลขคำสั่งซื้อ': 'shopee_no',
    'รหัสคืนสินค้า': 'return',
    'สินค้าราคาปกติ': 'amount_full',
    'จำนวนเงินทั้งหมดที่โอนแล้ว (฿)': 'after_fee'
})

df = df[df["return"].isna()]

df_shopee_cleaned = df[["date", "shopee_no", "amount_full", "after_fee"]]
```

```python
df_shopee_cleaned
```

```python
df_simas = data["raw_hq_simas_sales_bills.csv"].copy()
```

```python
df_out = df_shopee_cleaned.merge(
    df_simas[["PO", "BILLNO"]],
    how="left",
    left_on="shopee_no",
    right_on="PO"
)
```

```python
df_out["deduction_pct"] = (
    (df_out["amount_full"] - df_out["after_fee"]) / df_out["amount_full"] * 100
).round(2)
```
